### Studio Covarianza

```python
%pylab inline
import pandas as pd
import os
```

Importo i dati:

```python
address_attesa = os.path.abspath(os.getcwd())+'/INDIPENDENZA DELLE RIVOLTE - presa dati2 - PANDAS1.csv'
address_partecipanti = os.path.abspath(os.getcwd())+'/INDIPENDENZA DELLE RIVOLTE - presa dati2 - PANDAS2.csv'
address_durata = os.path.abspath(os.getcwd())+'/INDIPENDENZA DELLE RIVOLTE - presa dati2 - PANDAS3.csv'
```

```python
dati_attesa = pd.read_csv(address_attesa, nrows=20).drop(['run n°'], axis=1)
dati_partecipanti = pd.read_csv(address_partecipanti, nrows=20).drop(['run n°'], axis=1)
dati_durata = pd.read_csv(address_durata, nrows=20).drop(['run n°'], axis=1)
```

Correlazione tra dati dello stesso tipo a ordini diversi

```python
def color_(val):
    '''
    Takes a scalar and returns a string with
    the css property `'color: red'` for
    strings with r>.2, black otherwise.
    '''
    color = 'red' if val > 0.2 else 'black'
    return 'color: %s' % color
```

```python
dati_attesa.corr().style.applymap(color_)
```

```python
dati_partecipanti.corr().style.applymap(color_)
```

```python
dati_durata.corr().style.applymap(color_)
```

```python
df = dati_attesa.join(dati_partecipanti).join(dati_durata)
df.head()
```

```python
df.corr()
```

```python
df.corr().style.applymap(color_)
```
